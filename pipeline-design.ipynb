# Task 1: Pipeline Architecture Diagram

## 1.1 End-to-End Pipeline Architecture

The data pipeline begins with two primary data sources. The first source is a historical batch file containing past transaction data, usually stored in formats such as CSV or XLSX. The second source is a live transaction stream that sends real-time records as they occur in the system.

Both data sources send their data to the ingestion layer. The ingestion layer is responsible for collecting incoming data and preparing it for storage and processing. Batch files can be ingested using scheduled workflows, while live transactions are captured through streaming systems.

After ingestion, the data is placed in a landing zone, also known as raw storage. The landing zone stores data in its original form without modification. This ensures that the raw data remains available for auditing, debugging, or reprocessing if necessary.

Next, the data passes through a validation stage. In this stage, records are checked for correctness. Validation checks include schema validation, data type checks, missing values, and logical constraints. This step ensures that only reliable data continues through the pipeline.

If a record fails validation, it is moved to a quarantine or dead letter area. This area stores problematic records so they can be analyzed and corrected later. Keeping invalid records separate prevents them from corrupting the main dataset.

Validated records move to the transformation stage. In this stage, the data is cleaned, standardized, and enriched. Common operations include removing duplicates, correcting inconsistent values, and generating additional features that will later be useful for analytics or machine learning.

After transformation, the data is stored in multiple storage layers. The raw layer stores the original unmodified data. The clean layer stores validated and processed data that is ready for analysis. The feature layer contains engineered features prepared specifically for machine learning models.

Once the data is stored, it becomes available to consumer interfaces. Business intelligence dashboards access the clean data layer to generate reports and visualizations. Machine learning models access the feature layer to train models and perform predictions.

Finally, monitoring is applied throughout the pipeline. Monitoring tools track data quality, pipeline performance, and system failures. If anomalies or failures are detected, alerts are sent to operators so they can investigate and resolve the issue quickly.

## 1.2 Component Descriptions

### Data Sources
The pipeline starts with two types of data sources: historical batch data and real-time transaction streams. The batch data typically comes from files such as CSV or Excel files containing historical retail transactions. The live stream represents real-time purchase events generated by an online retail system. These sources provide the raw input data for the entire pipeline.

### Ingestion Layer
The ingestion layer collects data from both batch files and live streams and brings it into the pipeline. It ensures that incoming data is properly formatted and transferred to the storage system. Batch ingestion may run on scheduled jobs, while streaming ingestion continuously processes incoming events.

### Landing Zone (Raw Storage)
The landing zone stores the ingested data exactly as it arrives from the source systems. No transformations are applied at this stage. The main purpose of this layer is to preserve the original data so it can be reprocessed or audited later if needed.

### Validation Stage
The validation stage checks whether the incoming data meets required structural and logical conditions. It verifies schema correctness, required fields, and acceptable value ranges. This step helps prevent corrupted or invalid records from entering the rest of the pipeline.

### Quarantine / Dead Letter Area
Records that fail validation are moved to a quarantine or dead letter area. This storage area isolates problematic records from valid data. Operators can later inspect these records, determine the cause of the issue, and correct the source data.

### Transformation Stage
The transformation stage cleans and prepares the validated data for analysis and machine learning. Typical operations include removing duplicates, fixing inconsistent formats, aggregating values, and generating new features. These transformations improve the usability and quality of the data.

### Storage Layers
The pipeline maintains multiple storage layers. The raw layer contains the original ingested data. The clean layer contains validated and processed data that is ready for analytics. The feature layer stores datasets that have been prepared specifically for machine learning models.

### Consumer Interfaces
Consumer systems access the processed data for different purposes. Business intelligence dashboards use the clean layer to generate reports and insights. Machine learning systems use the feature layer to train models and produce predictions.

### Monitoring
Monitoring tools track the health and performance of the pipeline. They monitor data quality metrics, processing times, and system errors. If problems occur, alerts are automatically sent to operators so they can investigate and resolve the issue quickly.

# Task 2: Validation and Error Handling Design

## 2.1 Validation Rules

### A. Schema Validations (Structural Correctness)

1. `InvoiceNo` must be a non-empty string.  
2. `StockCode` must be a non-empty string and cannot contain only spaces.  
3. `Description` can be null, but if present, it must be a string.  
4. `Quantity` must be an integer and cannot be null.  
5. `InvoiceDate` must be a valid datetime in the format `YYYY-MM-DD HH:MM:SS`.  
6. `UnitPrice` must be numeric and cannot be null.  
7. `CustomerID` can be null for anonymous customers, but if present, it must be non-empty.  
8. `Country` must be a valid ISO country name.

---

### B. Value Range Validations (Sensible Values)

1. `Quantity` cannot be zero.  
2. `Quantity` must be positive for non-cancellation records; negative values are allowed only if `InvoiceNo` starts with `"C"`.  
3. `UnitPrice` must be greater than or equal to 0.  
4. `UnitPrice` must be less than 10,000 to avoid extreme outliers.  
5. `InvoiceDate` cannot be in the future.  
6. `CustomerID`, if present, must be numeric.

---

### C. Business Rule Validations (Domain Logic)

1. If `InvoiceNo` starts with `"C"`, then `Quantity` must be negative.  
2. The total transaction amount (`Quantity * UnitPrice`) must be greater than or equal to 0.  
3. Customers from the UK must have a non-null `CustomerID`.  
4. Each combination of `InvoiceNo` and `StockCode` must be unique per day.  
5. `StockCode` must exist in the product master table.  
6. The `Description` of `StockCode` must match the master table entry.

## 2.2 Error Handling Flow

### Schema Validation Failures
- **Action:** The record is rejected entirely.  
- **Destination:** Dead letter queue or quarantine table.  
- **Operator Alert:** Email or Slack notifications with a summary of schema errors.  
- **Recovery:** After fixing the source or ETL mapping, quarantined records can be reprocessed.

### Value Range Validation Failures
- **Action:** The record is flagged for review; partially accepted if non-critical.  
- **Destination:** Quarantine table with detailed reason for failure.  
- **Operator Alert:** Monitoring dashboards display error counts; critical violations trigger Slack/email alerts.  
- **Recovery:** Correct the source values or manually fix the records in the quarantine table, then reprocess.

### Business Rule Validation Failures
- **Action:** The record is flagged for operator review and may be blocked from ML or BI consumption until resolved.  
- **Destination:** Quarantine table with a description of the violation.  
- **Operator Alert:** Automated alerts to data stewards or business analysts.  
- **Recovery:** Correct inconsistencies (e.g., adjust `Quantity` for cancellations, update master tables), then re-ingest the corrected records.

# Task 3: Transformation and Storage Design

## 3.1 Define Transformations

### Cleaning Operations

**Remove duplicate transactions**

Input: Validated transaction records from the clean input stream.  
Output: Dataset where duplicate records with the same InvoiceNo, StockCode, and InvoiceDate are removed.  
Idempotency: Yes. Running the operation multiple times will always produce the same result because duplicates are identified deterministically.

**Handle missing product descriptions**

Input: Records where the Description field may be null.  
Output: Missing descriptions are filled using the product master table when available, otherwise replaced with "Unknown Product".  
Idempotency: Yes. Once missing values are filled, running the operation again does not change the result.

**Standardize country names**

Input: Transaction records containing the Country field.  
Output: Country names standardized to a consistent format (for example, "UK" converted to "United Kingdom").  
Idempotency: Yes. Standardization rules always produce the same output.

---

### Derived Columns

**Calculate line total**

Input: Quantity and UnitPrice fields.  
Output: A new column called LineTotal calculated as Quantity × UnitPrice.  
Idempotency: Yes. The same input values always produce the same result.

**Create cancellation flag**

Input: InvoiceNo field.  
Output: A new boolean column called IsCancellation which is True if InvoiceNo starts with "C", otherwise False.  
Idempotency: Yes. The logic is deterministic.

**Extract date components**

Input: InvoiceDate timestamp.  
Output: New columns including InvoiceYear, InvoiceMonth, InvoiceDay, and InvoiceWeek.  
Idempotency: Yes. Date extraction produces consistent results.

---

### Customer-Level Aggregations

**Calculate total customer revenue**

Input: Transaction-level data grouped by CustomerID.  
Output: A new aggregated dataset containing TotalRevenue per customer.  
Idempotency: Yes. Aggregations produce the same results when applied to the same input dataset.

**Calculate order count per customer**

Input: Transaction data grouped by CustomerID and InvoiceNo.  
Output: Number of unique orders per customer stored as OrderCount.  
Idempotency: Yes.

**Calculate product diversity**

Input: Transactions grouped by CustomerID and StockCode.  
Output: Number of distinct products purchased by each customer.  
Idempotency: Yes.

**Calculate recency**

Input: InvoiceDate and CustomerID.  
Output: Recency value representing the number of days since the customer's last purchase.  
Idempotency: Yes, as long as the observation date is fixed.

---

### Feature Engineering for the ML Model

**Customer lifetime value feature**

Input: Aggregated transaction data for each customer.  
Output: Estimated customer lifetime value based on historical spending patterns.  
Idempotency: Yes.

**Average order value**

Input: TotalRevenue and OrderCount per customer.  
Output: A new feature called AvgOrderValue calculated as TotalRevenue divided by OrderCount.  
Idempotency: Yes.

**Purchase frequency**

Input: Number of orders and observation window length.  
Output: Feature representing how frequently a customer purchases during the observation window.  
Idempotency: Yes.

## 3.2 Design the Storage Layers

| Layer   | Contents | Format | Update Frequency | Retention |
|--------|---------|--------|------------------|-----------|
| Raw | Original ingested transaction data from batch and streaming sources | Parquet / Delta | Continuous ingestion | Long-term storage |
| Clean | Validated and transformed transaction data | Parquet / Delta | Daily updates | Medium-term retention |
| Feature | Customer-level features prepared for machine learning models | Parquet / Delta | Daily or weekly refresh | Based on model training needs |

### Format Justification

**Raw Layer**

The raw layer stores the original ingested data exactly as it arrives from the source systems. Parquet or Delta format is chosen because it supports efficient compression and large-scale storage. This layer may grow very large over time because it stores all historical data. It is mainly used for auditing, debugging, and reprocessing pipelines.

**Clean Layer**

The clean layer contains validated and transformed data that is ready for analytics. Analysts and BI tools frequently read from this layer to perform reporting and analysis. Parquet or Delta format enables fast analytical queries and efficient storage for large datasets.

**Feature Layer**

The feature layer stores engineered features used by machine learning models. It must support efficient reads for training and inference workflows. Delta Lake is beneficial because it supports versioning and time travel, which allows models to reproduce historical training datasets.

---

## 3.3 Incremental Update Strategy

### Tracking Processed Data

The pipeline tracks processed records using a high-water mark strategy based on the InvoiceDate or ingestion timestamp. Each processing job records the latest processed timestamp so that the next run only processes new data that arrives after this point.

### Handling Late-Arriving Records

Late-arriving records are handled by allowing a small processing window where recent data can still be updated. If a record arrives late but falls within the allowed window, the pipeline updates the affected aggregates and features.

### Refreshing Customer-Level Features

Customer-level features are refreshed on a scheduled basis, typically once per day. The pipeline recomputes aggregated metrics such as revenue, order count, and recency using the latest available data. Machine learning feature datasets are updated after these aggregations to ensure that models always use the most recent customer behavior data.
